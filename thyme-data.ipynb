```python
from datasets import load_dataset
import base64

# Load dataset in streaming mode (doesn't download everything)
dataset = load_dataset("Kwai-Keye/Thyme-RL", split="train", streaming=True)

# Fetch first 5 samples
samples = []
for i, sample in enumerate(dataset):
    if i >= 20:
        break
    samples.append(sample)
    print(f"Sample {i+1}:")
    print(sample)
    print("\n" + "="*80 + "\n")
```

```python
samples[0]
```

```python
from PIL import Image
import io
import base64
import json


for i, sample in enumerate(samples):
    image_string = sample['images'][0]  # This is a base64 string
    
    # Decode base64 string to bytes
    image_bytes = base64.b64decode(image_string)
    
    # Now open with PIL
    image = Image.open(io.BytesIO(image_bytes))
    image.save(f"thyme_sft/sample_{i+1}_image.png")
    print(f" Saved sample_{i+1}_image.png - Size: {image.size}")

    # save the questions and answers to a text file and messages to a json file
    sample_json = {
        "messages": sample['messages'],
        "question": sample['question'],
        "solution": sample['solution']

    }

    with open(f"thyme_rl/sample_{i+1}_data.json", "w") as f:
        json.dump(sample_json, f, indent=4)
        
```
